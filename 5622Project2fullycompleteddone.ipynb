# 2. AlexNet model setting

```python
import torch
import torchvision.models as models
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import DataLoader
import os



# 加载预训练的AlexNet模型
model = models.alexnet(pretrained=True)

# 打印模型的结构
#print(model)

# 冻结前面的卷积层
for param in model.features.parameters():
  param.requires_grad = False

# 替换输出层以适应新任务
num_classes = 1000  # 你的数据集中的类别数
model.classifier[6] = nn.Linear(4096, num_classes)

customized_linear_layer_1 = nn.Linear(num_classes, 100)
customized_linear_layer_2 = nn.Linear(100, 20)
customized_linear_layer_3 = nn.Linear(20, 10)

custom_layer = nn.Sequential(
  customized_linear_layer_1,
  nn.ReLU(),
  nn.Dropout(0.25),
  customized_linear_layer_2,
  nn.ReLU(),
  nn.Dropout(0.25),
  customized_linear_layer_3,
  nn.ReLU(),
  nn.Dropout(0.25)
)

model.custom_layers = custom_layer


# 对于模型的每个权重，使其不进行反向传播，即固定参数
for param in model.parameters():
    param.requires_grad = False

# 但是参数全部固定了，也没法进行学习，所以我们不固定最后一层，即全连接层
for param in model.features[10].parameters():
    param.requires_grad = True
for param in model.classifier[6].parameters():
    param.requires_grad = True
for param in model.custom_layers.parameters():
    param.requires_grad = True

#print(model)


# 指定要创建的文件夹的路径
folder_path = "/content/best_model"

# 使用os.makedirs()创建文件夹，如果文件夹已存在，也不会引发错误
os.makedirs(folder_path, exist_ok=True)

```

```python
import torch
import torchvision.models as models
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import DataLoader
import random

#与最终一致test
def validationset_precision(net, testset):


    net.eval()
    total_count = 0
    total_correct = 0
    total_loss = 0
    #count = 0
    validation_loss = 0
    criterion = nn.CrossEntropyLoss()
    total_samples = 0

    with torch.no_grad():
      dl = DataLoader(testset, batch_size=128, drop_last=True)
      #rotate function
      for i in dl:
        for j in i[0]:
          rotation_angle = random.randint(-180, 180)
          random_rotation = transforms.RandomRotation(degrees=(rotation_angle, rotation_angle))
          rotated_tensor = random_rotation(j)

      for data in dl:
          inputs = data[0].cuda()         #shape(128x3x32x32)
          targets = data[1].cuda()        #shape(128)
          #print("input")
          #print(type(inputs))
          outputs = net(inputs)
          loss = criterion(outputs, targets)
          #num = loss.item()
          validation_loss += loss.item()
          total_samples += targets.size(0)

          #validation_loss = model.evaluate(validation_dataset)
          #total_loss = total_loss + loss
          #print(count)
          #print(outputs.shape)
          #print(targets)
          #print(targets.shape)
          #print(outputs)
          predicted_labels = outputs.argmax(dim=1)
          #print(predicted_labels)
          #print(targets)
          comparison = predicted_labels == targets
          total_count += predicted_labels.size(0)
          total_correct += comparison.sum()
          #print("total correct")
          #print(total_correct)
    average_validation_loss = validation_loss / total_samples
    net.train()

    return int(total_correct) / int(total_count), average_validation_loss

print(f'Inital precision: {validationset_precision(model, validation_dataset)}')

from IPython import display
import matplotlib.pyplot as plt
class DrawingBoard:
    def __init__(self, names):
        self.data = {}
        for name in names:
            self.data[name] = []

    def update(self, data_dict):
        for key in data_dict:
            self.data[key].append(data_dict[key])

    def draw(self):
        all_keys = list(self.data.keys())
        fig, ax = plt.subplots(nrows=1, ncols=len(all_keys)-1)
        for idx in range(len(all_keys)-1):
            if idx == 0:
              ax[idx].plot(self.data[all_keys[0]], label="training loss")
              ax[idx].plot(self.data[all_keys[1]], label="validation loss")
              ax[idx].set_title(all_keys[0])
              ax[idx].legend()
            else:
              ax[idx].plot(self.data[all_keys[2]])
              ax[idx].set_title(all_keys[2])
        display.clear_output(wait=True)
        plt.show()
```

```python
import torch
import torchvision.models as models
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import DataLoader
import os

print('start training')
db = DrawingBoard(['training_loss', 'validationset_loss', 'validationset_precision'])
trainLoader = DataLoader(train_dataset, batch_size=200)
#rotate function
for i in trainLoader:
  for j in i[0]:
    rotation_angle = random.randint(-180, 180)
    random_rotation = transforms.RandomRotation(degrees=(rotation_angle, rotation_angle))
    rotated_tensor = random_rotation(j)

best_model = None
best_validation_loss = 1
no_useful_epoch = 0
max_epoch = 30


for epoch in range(300):
  total_samples = 0
  train_loss = 0
  for data in trainLoader:
    if epoch < 150:
      inputs = data[0].cuda()
      labels = data[1].cuda()
      optimizer1.zero_grad()
      outputs = model(inputs)
      loss = criterion(outputs, labels)
      train_loss += loss.item()
      total_samples += labels.size(0)
      #print(loss)
      loss.backward()
      optimizer1.step()
    else:
      inputs = data[0].cuda()
      labels = data[1].cuda()
      optimizer2.zero_grad()
      outputs = model(inputs)
      loss = criterion(outputs, labels)
      train_loss += loss.item()
      total_samples += labels.size(0)
      #print(loss)
      loss.backward()
      optimizer2.step()

  average_train_loss = train_loss / total_samples
  #print(validationset_precision(model, validation_dataset))
  vali_precision, vali_loss = validationset_precision(model, validation_dataset)
  #print(vali_loss)

  db.update({'training_loss': average_train_loss, 'validationset_loss': vali_loss, 'validationset_precision': vali_precision})
  db.draw()

  if vali_loss < best_validation_loss:
        best_validation_loss = vali_loss
        #os.remove('/content/best_model')
        model_state_dict = model.state_dict()
        torch.save(model_state_dict, '/content/best_model/model.pt')

        no_useful_epoch = 0
  else:
      no_useful_epoch += 1

  # 提前停止训练
  if no_useful_epoch >= max_epoch:
      print('Get the optimal model in training')
      break


print('finished')
```
